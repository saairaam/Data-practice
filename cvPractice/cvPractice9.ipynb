#Thresholding


```python
import cv2
import matplotlib.pyplot as plt
img1 = cv2.imread('../assest/waterbodies/waterbody2.png',0)
plt.imshow(img1)
```

```python
img1.max()
```

```python
ret,thres1=cv2.threshold(img1,127,255,cv2.THRESH_BINARY)
```

```python
ret
```

```python
plt.imshow(thres1,cmap='gray')
```

```python

thres2=cv2.threshold(img1,127,255,cv2.THRESH_BINARY)
plt.imshow(thres2[1], cmap='gray')
```

```python
thres3=cv2.threshold(img1,127,255,cv2.THRESH_TRUNC)
plt.imshow(thres3[1], cmap='gray')
```

```python
thres4=cv2.threshold(img1,127,255,cv2.THRESH_TRIANGLE)
plt.imshow(thres4[1], cmap='gray')
```

```python
thres5=cv2.threshold(img1,127,255,cv2.THRESH_OTSU)
plt.imshow(thres5[1], cmap='gray')
```

```python
thres6=cv2.threshold(img1,127,255,cv2.THRESH_TOZERO)
plt.imshow(thres6[1], cmap='gray')
```

```python
thres7=cv2.threshold(img1,127,255,cv2.THRESH_TOZERO_INV)
plt.imshow(thres7[1], cmap='gray')
```
