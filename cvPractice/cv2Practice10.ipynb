```python
import cv2
import matplotlib.pyplot as plt
```

```python
img= cv2.imread('../assest/practice/crossword.jpg',0)
plt.imshow(img)
```

```python
def showImg(img):
    fig=plt.figure(figsize=(15,15))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
showImg(img)
```

```python
ret,thres1= cv2.threshold(img,127,255,cv2.THRESH_BINARY)
showImg(thres1)
```

```python
ret,thres2= cv2.threshold(img,170,255,cv2.THRESH_BINARY)
showImg(thres2)
```

```python
thres2= cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,8)
showImg(thres2)
```

```python
blendImage = cv2.addWeighted(src1=thres1,alpha=0.6,src2=thres2,beta=0.4,gamma=0)
showImg(blendImage)
```
