```python
import matplotlib.pyplot as plt
import cv2
# Load the image
img = cv2.imread('../assest/waterbodies/waterbody2.png')
# Sobel
sobel = cv2.Sobel(img, cv2.CV_64F, 1, 1, ksize=5)
# Laplacian
laplacian = cv2.Laplacian(img, cv2.CV_64F)
# Canny
edges = cv2.Canny(img, 100, 200)
# Plotting all
plt.subplot(1, 3, 1), plt.imshow(sobel, cmap='gray')
plt.title('Sobel'), plt.xticks([]), plt.yticks([])
plt.subplot(1, 3, 2), plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(1, 3, 3), plt.imshow(edges, cmap='gray')
plt.title('Canny'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python

sobel_y = cv2.Sobel(img, cv2.CV_64F, 1, 1, ksize=5)
# Combine the two
plt.subplot(1, 3, 1), plt.imshow(sobel_y)
plt.title('Sobel'), plt.xticks([]), plt.yticks([])
```
